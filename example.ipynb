# Check Memory Usage


```python
import psutil

THRESHOLD_GB = 3 # Memory usage threshold in GB

def check_memory_usage():
    mem = psutil.virtual_memory()
    current_gb = mem.used / (1024**3)
    if current_gb > THRESHOLD_GB:
        print(f"Warning: High memory usage detected! Current usage is {current_gb:.2f} GB")

check_memory_usage()
```

# Send message to Telegram

```python
import requests

BOT_TOKEN = "your_bot_token_here"
USER_ID = "your_user_id_here"

def send_telegram(msg):
    url = f"https://api.telegram.org/bot{BOT_TOKEN}/sendMessage"
    params = {"chat_id": USER_ID, "text": msg}
    requests.get(url, params=params)
    
send_telegram("Hello from Python script!")
```
